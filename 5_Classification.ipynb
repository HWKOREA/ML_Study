```python
# 지도 학습 (Supervised Learning)

# 입출력 쌍의 예시들을 기반으로 입력을 출력으로 맵핑하는 함수를 학습하는 기계 학습 방법
# 함수는 라벨이 붙어 있는 학습 데이터 (labeled training data) 로부터 추론된다.
# 입력과 원하는 출력을 모두 포함하는 데이터 세트의 수학적 모델을 구축한다.
# 적용 분야 : 분류, 회귀 분석 등
# 입력 변수가 존재하며 측정되거나 사전에 준비된다.
# 입력 변수는 하나 혹은 그 이상의 출력에 영향을 준다.
# 목표는 입력을 사용해출력 값을 예측하는 것이다.
# 입력
# - 예측변수 (Predictor)
# - 독립변수 (Independent Variable)
# - 특성 / 특징 (Feature)
# 출력
# - 반응 (Response)
# - 종속변수 (Dependant Variable)
```

![Alt text](image-15.png)

```python
# 변수 타입과 용어

# 양적 (Quantitative) 변수
# - 포도당 예측 예제

# 질적 (Qualitative) 변수
# - 붓꽃 판별 예제 : G = {Virginia, Setosa, Versicolor}
# - 손글씨 숫자 예제 : G = {0, 1, ..., 9}
# - 클래스에서 명시적인 순서가 없다.
# - 클래스를 표시하는 데 숫자보다는 기술적 라벨 (Descriptive Label) 이 더 자주 사용된다.
# - 다른 말로 범주형 (Categorical) 변수, 이산 (Discrete) 변수, 요인 (Factor)

# 입력을 사용해 출력을 예측
# - 오늘과 어제의 특정한 대기 측정치 -> 내일의 오존 수준 예측
# - 손글씨 숫자의 회색조 값 -> 클래스 라벨 예측

# 출력 타입의 차이
# - 양적 출력 : 회귀 (Regression)
# - 질적 출력 : 분류 (Classification)
# - 비슷한 점 : 함수 근사적 과제로 볼 수 있다.

# 입력 또한 측정치 타입에 따라 다르다.

# 세 번째 변수 타입
# - Small, Medium, Large 와 같이 순서 있는 범주형 (Ordered Categorical)
# - 값 사이에 순서가 있지만 적절한 측량적 개념이 없다.

# 질적 변수
# - 코드에 의해 수치적으로 표현된다.
# - 목표 (Target) 라고 부르기도 한다.
# - 성공과 실패 -> 0과 1, -1과 1의 단일 이진 숫자나 비트로 표현
# - 두 개 이상의 범주 : 몇 가지 대안이 존재

# 더미 변수 (Dummy Variable) 사용
# - K 수준의 질적 변수가 K개의 이진변수 또는 비트의 벡터로 표현되어 있으며 이 중 오직 하나만 ON

# 입력 변수를 기호 X로 표시
# - X가 벡터라면 구성요소는 Xj로 표시

# 양적 출력은 Y로 표시

# 질적 출력은 G (그룹)으로 표시

# 관측된 값은 소문자로 표시
# - X의 i번째 관측된 값 : xi

# 행렬은 굵은 대문자

# 학습 과제 예시
# - 입력 벡터 X의 값이 주어졌을 때 Y_hat 이라 표기하는 출력 Y의 좋은 예측을 하라

# 두 개 클래스 G에 관한 한 가지 접근법
# - 이진으로 코드된 목표를 Y라고 표기하고 양적인 출력으로 다룬다.
# - 예측 Y_hat은 [0, 1] 안에 포함될 것이며 y_hat > 0.5 인지에 따라 G_hat 클래스 라벨에 할당할 수 있다.

# 훈련 데이터 (Training Data)
# - 사용가능한 관측치 집합 (xi, yi) 또는 (xi, gi), i = 1, ..., N
```

```python
# 결정 트리의 특징

# 장점
# - 알고리즘이 쉽고 직관적
# - 룰이 명확하고 시각화가 쉬움
# - 각 피처의 스케일링, 정규화 등의 전처리 작업 필요 없음

# 단점
# - 과적합으로 정확도가 떨어짐
# - 정확도를 높이기 위해 계속 조건을 추가하면 트리가 깊고 복잡해짐
#   -> 테스트 데이터에 유연하게 대처하지 못함
```

```python
# DecisionTreeClassifier 파라미터

# min_samples_split
# - 노드를 분할하기 위한 최소한의 샘플 데이터 수 (기본 값: 2)
# min_samples_leaf
# - 말단 노드가 되기 위한 최소한의 샘플 데이터 수
# max_features
# - 최대 피처 수 (기본 값: None, 모든 피처 사용)
# - int 값   : 피처의 개수
# - float 값 : 퍼센트
# - ‘sqrt’   : √(전체 피처 개수)
# - ‘auto’   : ‘sqrt’와 같음
# - ‘log’    : log_2⁡〖(전체 피처 개수)〗
# - ‘None’   : 전체 피처
# max_depth
# - 트리의 최대 깊이
# - 기본 값 None :
#   클래스 결정될 때까지 또는 노드의 데이터 개수가 min_samples_split보다 작을 때까지 깊이 증가
# max_leaf_nodes
# - 말단 노드의 최대 개수
```

```python
# 결정 트리 (실행이 안됨요........)

from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
import matplotlib.pyplot as plt

iris = load_iris()
x_train, x_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size = 0.2)
clf = DecisionTreeClassifier()
clf.fit(x_train, y_train)

plt.figure(figsize=(16, 16))
tree.plot_tree(clf, feature_names = iris.feature_names, class_names = [str(target) for target in iris.target_names], filled=True)
```

```python
clf = DecisionTreeClassifier(max_depth = 3)
clf.fit(x_train, y_train)

plt.figure(figsize=(16, 16))
tree.plot_tree(clf, feature_names = iris.feature_names, class_names = [str(target) for target in iris.target_names], filled=True)
```

```python
clf = DecisionTreeClassifier(min_samples_split = 4)
clf.fit(x_train, y_train)

plt.figure(figsize=(16, 16))
tree.plot_tree(clf, feature_names = iris.feature_names, class_names = [str(target) for target in iris.target_names], filled=True)
```

```python
clf = DecisionTreeClassifier(min_samples_leaf = 4)
clf.fit(x_train, y_train)

plt.figure(figsize=(16, 16))
tree.plot_tree(clf, feature_names = iris.feature_names, class_names = [str(target) for target in iris.target_names], filled=True)
```

```python
# 피처의 중요도

# DecisionTreeClassifier
# - feature_importances_ 속성

print(clf.feature_importances_)
plt.barh(iris.feature_names, clf.feature_importances_)
```

```python
# 결정 트리 과적합 (Overfitting)

from sklearn.datasets import make_classification

x, y = make_classification(n_features = 2, n_redundant = 0, n_classes = 3, n_clusters_per_class = 1)
plt.scatter(x[:, 0], x[:, 1], c = y)
```

```python
import numpy as np

def visualize(model, x, y):
    x1 = x[:, 0]
    x2 = x[:, 1]
    plt.scatter(x1, x2, c = y)
    x1_min = np.min(x1)
    x1_max = np.max(x1)
    x2_min = np.min(x2)
    x2_max = np.max(x2)
    xx, yy = np.meshgrid(np.linspace(x1_min, x1_max, num = 200), np.linspace(x2_min, x2_max, num = 200))
    z = model.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)
    n_classes = len(np.unique(y))
    plt.contourf(xx, yy, z, alpha = 0.3, cmap = 'rainbow')
    
clf = DecisionTreeClassifier()
clf.fit(x, y)
visualize(clf, x, y)
```

```python
clf = DecisionTreeClassifier(min_samples_leaf = 6)
clf.fit(x, y)
visualize(clf, x, y)
```

```python
# 사용자 행동 인식

import pandas as pd

df = pd.read_csv('C:/woogi/KWU/3-2/Machine_Learning/uci/features.txt',sep = '\s+', header = None, names = ['column_index', 'column_name'])
df
```

```python
# 중복된 피처 확인

dup = df.groupby('column_name').count()
print(dup[dup['column_index'] > 1].count())
dup[dup['column_index'] > 1].head()
```

```python
# 중복된 피처 처리

df2 = pd.DataFrame(data = df.groupby('column_name').cumcount(), columns = ['cnt'])
df2 = df2.reset_index()
df3 = pd.merge(df.reset_index(), df2, how = 'outer')
df3['column_name'] = df3[['column_name', 'cnt']].apply(\
    lambda x : x[0] + '_' + str(x[1]) if x[1] > 0 else x[0], axis = 1)
df3 = df3.drop(['index'], axis = 1)
feature_name = df3.iloc[:, 1].values.tolist()
```

```python
# 데이터 읽기

X_train = pd.read_csv('C:/woogi/KWU/3-2/Machine_Learning/uci/train/X_train.txt', sep = '\s+', names = feature_name)
X_test = pd.read_csv('C:/woogi/KWU/3-2/Machine_Learning/uci/test/X_test.txt', sep = '\s+', names = feature_name)
y_train = pd.read_csv('C:/woogi/KWU/3-2/Machine_Learning/uci/train/y_train.txt', sep = '\s+', header = None, names = ['action'])
y_test = pd.read_csv('C:/woogi/KWU/3-2/Machine_Learning/uci/test/y_test.txt', sep = '\s+', header = None, names = ['action'])
```

```python
print(X_train.info())
```

```python
print(y_train['action'].value_counts())
```

```python
# 하이퍼 파라미터 값들

from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

clf = DecisionTreeClassifier()
clf.fit(X_train, y_train)
y_hat = clf.predict(X_test)
accuracy = accuracy_score(y_test, y_hat)
print(accuracy)

print(clf.get_params())
```

```python
from sklearn.model_selection import GridSearchCV

params = {'max_depth' : [6, 8, 10, 12, 16, 20, 24]}
cv = GridSearchCV(clf, param_grid = params, scoring = 'accuracy', cv = 5)
cv.fit(X_train, y_train)
print(cv.best_score_)
print(cv.best_params_)
```

```python
results = pd.DataFrame(cv.cv_results_)
results[['param_max_depth', 'mean_test_score']]
```

```python
params = {'max_depth' : [6, 8, 10, 12, 16, 20, 24], 'min_samples_split' : [16, 24]}
cv = GridSearchCV(clf, param_grid = params, scoring = 'accuracy', cv = 5)
cv.fit(X_train, y_train)
print(cv.best_score_)
print(cv.best_params_)
```

```python
best_clf = cv.best_estimator_
y_hat = best_clf.predict(X_test)
accuracy = accuracy_score(y_test, y_hat)
print(accuracy)
```
